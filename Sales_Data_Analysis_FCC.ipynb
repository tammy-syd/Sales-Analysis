# Data cleaning and Modification

## Delete null values

```python
# Check and delete null values

all_df.isnull().any()

# Drop rows with NaN

nan_df=all_df[all_df.isna().any(axis=1)]
nan_df
```

```python
all_df=all_df.dropna(how='all')
all_df
```

### Add month column
### Use to answer Q1

```python



all_df['Month']= all_df['Order Date'].str[0:2]

# Convert Month from str to int

#all_df['Month']= all_df['Month'].astype('int32')

# Find and delete 'Or' to prevent error
all_df=all_df[all_df['Order Date'].str[0:2] != 'Or']

all_df['Month']= all_df['Month'].astype('int32')


all_df
```

### Convert columns from str to int

```python
# New way
all_df['Quantity Ordered']= pd.to_numeric(all_df['Quantity Ordered'])

all_df['Price Each']= all_df['Price Each'].astype('float')

all_df
```

### Add sales column = Quantity Ordered * Price Each	

```python
all_df['Sales'] = all_df['Price Each'] * all_df['Quantity Ordered']
all_df
```

# Q4: What products are most often sold together?

## Plot twist: Products sold together have the same ID

### Create new df with duplicated Order IDs


```python
new_df = all_df[all_df['Order ID'].duplicated(keep=False)]
new_df
```

### Add new column: Grouped
### Join products with same ID

```python
new_df['Grouped'] = new_df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
new_df
```

```python
# Get rid of duplicated in Grouped column

new_df = new_df[['Order ID','Grouped']].drop_duplicates()

new_df
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in new_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
# Most common combo
for key,value in count.most_common(10):
    print(key,value)
```

# Q5: What product sold the most? Why do you think it sold the most?

```python
product_group_total = all_df.groupby('Product').sum()['Quantity Ordered']

product_group_total
```

```python
products = [product for product, df in all_df.groupby('Product')]

plt.bar(products,product_group_total,color='green')

plt.xticks(products,rotation='vertical',color='black')

plt.ylabel('Quantity Ordered')

plt.xlabel('Product')

plt.show()
```

```python
# Explain why AAA Batteries and AA Batteries sold the most

price = all_df.groupby('Product').mean()['Price Each']

price
```

### Add a second y-axis to the product chart above

```python
fig,axis1 = plt.subplots()

axis1.bar(products,product_group_total,color='green')

axis2 = axis1.twinx()
axis2.plot(products,price,color='blue')

axis1.set_xlabel('Product')
axis1.set_ylabel('Number of Orders')
axis2.set_ylabel ('Price')

# Fix the name
axis1.set_xticklabels(products,rotation='vertical' )

plt.show()
```
